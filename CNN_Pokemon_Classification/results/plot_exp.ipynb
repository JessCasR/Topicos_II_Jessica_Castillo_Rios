```python
import torch
import matplotlib.pyplot as plt

results_exp0 = torch.load("exp0_results.pth")
results_exp0.keys()

results_exp1 = torch.load("exp1_results.pth")
results_exp1.keys()

results_exp2 = torch.load("exp2_results.pth")
results_exp2.keys()
```

## Gr√°ficas Experimento 0

```python
train_loss = results_exp0["train_loss"]
test_loss = results_exp0["test_loss"]
train_acc = results_exp0["train_acc"]
test_acc = results_exp0["test_acc"]

epochs = range(1, len(train_loss) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, test_loss, label="Test Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("P√©rdida por √©poca ‚Äì Exp0 (TinyVGG)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(epochs, train_acc, label="Train Accuracy")
plt.plot(epochs, test_acc, label="Test Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Precisi√≥n por √©poca ‚Äì Exp0 (TinyVGG)")
plt.legend()
plt.grid(True)
plt.show()
```
